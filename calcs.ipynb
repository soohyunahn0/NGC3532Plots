```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('d.csv') #isochrone data
data2 = pd.read_csv('simbad.csv')
data6 = pd.read_csv('gaiadata.csv')
```

```python
v = data2['MagV'] # - (5 * np.log10(1/(0.001*data2['plx']))) + 5
bv = data2['MagB'] - data2['MagV']
plt.plot(bv, v, '.', color = 'blue')

isoa = data1['V']
isob = data1['B'] - data1['V']

plt.title("Color-Color Diagram and Isochrone")
plt.plot(isob, isoa, '-', color = 'red') #plot B-V versus V
plt.gca().invert_yaxis()
plt.xlabel('B-V')
plt.ylabel('V')
plt.show()
```

```python
v = data2['MagV'] - (5 * np.log10(1/(0.001*data2['plx']))) + 5

bv = data2['MagB'] - data2['MagV']
plt.plot(bv, v, '.', color = 'blue') # plot B-V versus the absolute magnitude
plt.title("HR Diagram of NGC 3532")
plt.gca().invert_yaxis()
plt.xlabel('B-V')
plt.ylabel('Absolute Magnitude')
plt.show()
```

```python
xaxis = pd.Series(data6['pmdec'])
yaxis = pd.Series(data6['pmra'])
plt.xlabel('Right Ascension')
plt.ylabel('Declination')
plt.title("Stars from NGC 3532")
plt.plot(xaxis, yaxis, '.', color = 'orange')
plt.show()
```
